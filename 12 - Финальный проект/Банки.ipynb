```python
# Зададим функцию с параметрами необходимого набора данных
def first_look(df):
    
    # Сделаем новый набор данных, в который будет добавлять искомые значения
    fl_df = pd.DataFrame(columns=list(df.columns))
    
    # Начнем итерировать по колонкам набора данных
    for column in df.columns:
        fl_df.loc[1, column] = df[column].nunique()
        fl_df.loc[2, column] = df[column].isna().sum()
        fl_df.loc[3, column] = round(df[column].isna().sum() / len(df[column]), 4) * 100
        fl_df.loc[4, column] = df[column].duplicated().sum()
        fl_df.loc[5, column] = round(df[column].duplicated().sum() / len(df[column]), 4) * 100
    
    # Переименуем индексы строк
    fl_df.index = ['Уникальные значения',
                   'Пропуски',
                   'Пропуски, %',
                   'Дубликаты',
                   'Дубликаты, %']
    
    fl_df = fl_df.rename(columns = dict(thesaurus))
    fl_df = fl_df.fillna('')
    
    # Выводим получившийся набор данных
    return display(fl_df)
```

## Исследовательский анализ данных

Для проведения исследовательского анализа зададим вспомогательные функции:

```python
def analysis(df, column):
    plt.figure(figsize=(12, 6))

    ax1=plt.subplot(1, 2, 1)
    ax2=plt.subplot(2, 2, 2)
    ax3=plt.subplot(2, 2, 4)

    # Создаем график распределения
    ax1=sns.histplot(data=df,
                     x=df[column],
                     bins=50,
                     stat='density',
                     common_norm=False,
                     hue='churn',
                     kde=True,
                     ax=ax1)
    ax1.set_title(f'Распределение по признаку {dict(thesaurus)[column]}', size=15)
    ax1.tick_params(labelsize=14)
    ax1.set_xlabel(dict(thesaurus)[column])
    ax1.set_ylabel('Плотность')
     
    ax2=sns.boxplot(x=column, data=df, notch=False, ax=ax2)
    ax2.tick_params(labelsize=14, 
                    labelbottom=False,
                    labeltop=True)
    ax2.set_xlabel(dict(thesaurus)[column])
    
    # Делаем таблицу
    props=dict(boxstyle='round', facecolor='#279EFF', alpha=0.5)
    desc=df[column].describe().astype('int')
    desc=pd.DataFrame(desc)
    ax3.axis('off')   
    ax3=plt.table(cellText=desc.values,
                  rowLabels=desc.index,
                  edges='closed',
                  in_layout=True,
                  loc='center',
                  rowLoc='center')
    
    ax3.auto_set_font_size(False)
    ax3.set_fontsize(14)
    ax3.scale(0.3, 1.8)
    
    plt.tight_layout()
    return plt.show()
```

```python
def visual(df, column):
    # Находим процент ушедших клиентов в зависимости от объектов собственности
    data_churn=data.query('churn==True')
    df=(data.pivot_table(index=column,
                         values='churn',
                         aggfunc=['sum', 'count']))
    df.columns=df.columns.droplevel(level=1)
    df['perc']=df['sum']/df['count']*100
    df['sum']=df['sum'].astype('int')
    df['count']=df['count'].astype('int')
    df['perc']=df['perc'].astype('int')
    
    # Строим графики визуализации
    fig,(ax1,ax2)=plt.subplots(1, 2, 
                               figsize=(12, 4),
                               gridspec_kw={'width_ratios':[2, 1]})
    
    # Построим столбчатую диаграмму:
    ax1=sns.barplot(data=df,
                    x=df.index,
                    y='perc',
                    palette='pastel',
                    ax=ax1)
    
    # Делаем подписи
    for bar in ax1.patches:
        ax1.annotate(format(bar.get_height(),'.0f'),
                           (bar.get_x() + bar.get_width() / 2,
                            bar.get_height()//2), ha='center', va='center',
                            size=14, xytext=(0, 10),
                            textcoords='offset points')
    ax1.set_xlabel(dict(thesaurus)[column])
    ax1.set_ylabel('Процент оттока клиентов')
    
    # Делаем таблицу
    ax2.axis('off')   
    ax2=plt.table(cellText=df.values,
                  colLabels=['Отток', 'Всего', 'Процент'],
                  rowLabels=df.index,
                  edges='closed',
                  in_layout=True,
                  loc='center',
                  rowLoc='center')
    
    ax2.auto_set_font_size(False)
    ax2.set_fontsize(14)
    ax2.scale(1, 2.5)
    
    ax1.set_title(f'Процент оттока клиентов в зависимости от {dict(thesaurus)[column]}', 
                  fontsize=14)
    plt.tight_layout(pad=0)
    
    # Строим графики визуализации
    fig,(ax3,ax4)=plt.subplots(1, 2, 
                               figsize=(12, 4),
                               gridspec_kw={'width_ratios':[2, 1]})
      
    # Построим столбчатую диаграмму:
    ax3=sns.barplot(data=df,
                    x=df.index,
                    y='sum',
                    palette='pastel',
                    ax=ax3)
    
    # Делаем подписи
    for bar in ax3.patches:
        ax3.annotate(format(bar.get_height(),'.0f'),
                           (bar.get_x() + bar.get_width() / 2,
                            bar.get_height()//2), ha='center', va='center',
                            size=14, xytext=(0, 10),
                            textcoords='offset points')
    ax3.set_xlabel(dict(thesaurus)[column])
    ax3.set_ylabel('Количество оттока клиентов')
    ax3.set_title(f'Количество оттока клиентов в зависимости от {dict(thesaurus)[column]}', 
                  fontsize=14)
    
    # Добавим категорию другие для значений меньше 500
    count=df['sum']
    count = count.groupby(np.where(count>=60,count.index,'Другие')).sum()

    ax4.pie(count, 
            labels = count.index,
            colors = sns.color_palette('pastel'),
            autopct ='%.0f%%')
    ax4.title.set_text(f'Соотношение оттока по категории {dict(thesaurus)[column]}')
    plt.tight_layout(pad=0)
    return plt.show()
```

## Проверка гипотез

Т-Тест предполагает, что данные:
   - независимы
   - распределены (приблизительно) нормально
   - имеют одинаковую величину дисперсии в каждой сравниваемой группе (т. н. однородность дисперсии)

Для проверки данных на равенство мы будем использовать t-test_ind:  
Это двусторонний тест для нулевой гипотезы о том, что две независимые выборки имеют идентичные средние(ожидаемые) значения.

### Гипотеза №1

Проверьте гипотезу различия возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним. 
Сформулируем *нулевую* и *альтернативную гипотезы*:  


**H0**: различие возраста клиентов, пользующихся одним продуктом банка = различию возраста клиентов, пользующихся двумя продуктами  

**Н1**: различие возраста клиентов, пользующихся одним продуктом банка ≠ различию возраста клиентов, пользующихся двумя продуктами

```python
# Формируем наборы данных
one_prod=data.loc[data['products']==1]['age']
two_prod=data.loc[data['products']==2]['age']

plt.figure(figsize=(8, 4))
sns.kdeplot(one_prod, 
            fill=True, 
            label='Один продукт', 
            common_norm=False)
sns.kdeplot(two_prod, 
            fill=True, 
            label='Два продукта', 
            common_norm=False)

plt.title('Распределение плотности по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.legend(loc = 1)

plt.show()
```

```python
results=st.ttest_ind(one_prod,
                     two_prod,
                     equal_var=False)

alpha=0.01

print(f'p-value = {results.pvalue}') 

if results.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

По результатам теста не получилось отвергнуть нулевую гипотезу, таким образом можно сказать, что различия возраста клиентов пользующихся 1 или 2 продуктами незначительны. 

Здесь стоит упомянуть, что был использован alpha = 0.1, т.е. довольно жесткая вероятность ошибочного отклонения истинной нулевой гипотезы. При alpha = 0.5 данную гипотезу можно было бы отвергать. 

### Гипотеза №2

Проверим гипотезу различия кредитного скоринга в разрезе отточности.
Сформулируем *нулевую* и *альтернативную гипотезы*:  


**H0**: рейтинговый скоринг у оставшихся клиентов = рейтинговому скорину у отточных клиентов

**Н1**: рейтинговый скоринг у оставшихся клиентов ≠ рейтинговому скорингу у отточных клиентов

```python
# Формируем наборы данных
churn_card=data.loc[data['churn']==0]['score']
client_card=data.loc[data['churn']==1]['score']

plt.figure(figsize=(8, 4))
sns.kdeplot(churn_card, 
            fill=True, 
            label='Отточный клиент', 
            common_norm=False)
sns.kdeplot(client_card, 
            fill=True, 
            label='Оставшийся клиент', 
            common_norm=False)

plt.title('Распределение плотности по рейтингу скоринга')
plt.xlabel('Рейтинг скоринга')
plt.ylabel('Плотность')
plt.legend(loc = 2)

plt.show()
```

```python
results=st.ttest_ind(churn_card,
                     client_card,
                     equal_var=False)

alpha=0.05

print(f'p-value = {results.pvalue}') 

if results.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

По результатам теста видно, что скоринговый рейтинг оставшихся и ушедших клиентов разный.
